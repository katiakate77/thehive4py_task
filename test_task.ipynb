```python
class CustomFieldHelper:
    def __init__(self):
        self._fields = {}

    def __add_field(self, type_, name, value):
        custom_field = {}
        custom_field[type_] = value
        self._fields[name] = custom_field

    def add_integer(self, name, value):
        self.__add_field('integer', name, value)
        return self

    def add_string(self, name, value):
        self.__add_field('string', name, value)
        return self

    def add_boolean(self, name, value):
        self.__add_field('bool', name, value)
        return self

    def build(self):
        return self._fields
```

```python
fields_dict = {
    'from': None,
    'category': 'gf',
    'incident-title': None,
    'incident_status': 4,
}

custom_fields_helper = CustomFieldHelper()

for name, value in fields_dict.items():
    if value is not None:
        custom_fields_helper.add_string(name, value)
custom_fields = custom_fields_helper.build()

custom_fields
```

```python
fields_dict = {
    'from': None,
    'category': 'gf',
    'incident-title': None,
    'incident_status': 4,
}

customFields = (CustomFieldHelper()
                .add_string('from', fields_dict['from'])
                .add_string('category', fields_dict['category'])
                .add_string('incident-title', fields_dict['incident-title'])
                .add_string('incident_status', fields_dict['incident_status']).build()
)
customFields
```
